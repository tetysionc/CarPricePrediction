# Wykresy


```
def filter_for_plot(feat_groupby,feat_agg='price_value', agg_funcs=[np.mean, np.median, np.size], feat_sort='mean', top=50, subplots=True):
  return(
     df
     .groupby(feat_groupby)[feat_agg]
     .agg(agg_funcs)
     .sort_values(by=feat_sort, ascending=False)
     .head(top) 
  ).plot(kind='bar', figsize=(15, 5), subplots = subplots)
```

```
filter_for_plot('param_co2-emissions');
```

```
filter_for_plot('param_rok-produkcji');
```
